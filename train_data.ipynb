```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load dataset
df = pd.read_csv(r"C:\Users\sarve\Desktop\credit_card_fruda\fraudTrain.csv")

# Parse datetime
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['dob'] = pd.to_datetime(df['dob'])

# Create time-based features
df['hour'] = df['trans_date_trans_time'].dt.hour
df['day'] = df['trans_date_trans_time'].dt.day
df['weekday'] = df['trans_date_trans_time'].dt.weekday

# Calculate age
df['age'] = df['trans_date_trans_time'].dt.year - df['dob'].dt.year

# Encode gender: 'M' -> 0, 'F' -> 1
df['gender'] = df['gender'].map({'M': 0, 'F': 1})

# Label encode category and merchant
le_cat = LabelEncoder()
df['category'] = le_cat.fit_transform(df['category'])

le_merchant = LabelEncoder()
df['merchant'] = le_merchant.fit_transform(df['merchant'])

# Drop columns you won't use
df = df.drop(columns=[
    'Unnamed: 0', 'first', 'last', 'street', 'city', 'state', 'zip',
    'job', 'dob', 'trans_num', 'cc_num', 'trans_date_trans_time'
])

# Define features for modeling
feature_cols = [
    'amt', 'category', 'merchant', 'gender', 'lat', 'long',
    'merch_lat', 'merch_long', 'city_pop', 'age',
    'hour', 'day', 'weekday'
]

# Scale numeric features
scaler = StandardScaler()
df[feature_cols] = scaler.fit_transform(df[feature_cols])

# Final dataset
X = df[feature_cols].values
y = df['is_fraud'].values

# View a preview
print(pd.DataFrame(X, columns=feature_cols).head())
print(pd.Series(y).value_counts())


print(X)
print(y)
```

```python
# from imblearn.over_sampling import SMOTE
# from imblearn.under_sampling import RandomUnderSampler
# from imblearn.under_sampling import TomekLinks
# from sklearn.model_selection import train_test_split
# import pandas as pd

# # Load your dataset (replace with your actual dataset)
# df = dfp

# # Separate features and labels
# X = df.drop('is_fraud', axis=1)
# y = df['is_fraud']

# # Define oversampling (SMOTE) and undersampling (Tomek Links)
# smote = SMOTE(random_state=42)
# tomek = TomekLinks()
# rus = RandomUnderSampler(random_state=42)

# # Apply SMOTE to oversample the minority class in the entire dataset
# #X_res_smote, y_res_smote = smote.fit_resample(X, y)

# # Now apply Tomek Links to clean the data
# X_res, y_res = tomek.fit_resample(X, y)
# X_rus,y_rus=rus.fit_resample(X_res,y_res)
# X_res, y_res = tomek.fit_resample(X_rus, y_rus)

# # Check the class distribution after both SMOTE and Tomek Links
# print("Original class distribution:")
# print(pd.Series(y).value_counts())

# print("Resampled class distribution:")
# print(pd.Series(y_res).value_counts())


import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import TomekLinks

# Load your preprocessed DataFrame
df = dfp

# Separate features and target
X = df.drop('is_fraud', axis=1)
y = df['is_fraud']

# Train-test split (important to avoid data leakage)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Step 1: Apply SMOTE to oversample minority class
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X_train, y_train)

# Step 2: Apply Tomek Links to clean borderline majority samples
tomek = TomekLinks()
X_res, y_res = tomek.fit_resample(X_smote, y_smote)

# Show class distributions
print("Original training set class distribution:")
print(y_train.value_counts())
print("\nAfter SMOTE + Tomek Links:")
print(pd.Series(y_res).value_counts())
```

```python
# Create a DataFrame of the resampled data
resampled_df = pd.DataFrame(X_res, columns=feature_cols)
resampled_df['is_fraud'] = y_res

# Show the first few rows
print(resampled_df.head())
```
